```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df.describe()
```

```python
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.jointplot(data=df,x='SepalLengthCm',y='SepalWidthCm',size=5,space=0.2,color='g')
```

```python
sns.FacetGrid(df,hue='Species',size=5)\
.map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
.add_legend()
```

```python
sns.boxplot(data=df,x='Species',y='PetalLengthCm')
```

```python
ax = sns.boxplot(data=df,x='Species',y='PetalLengthCm')
ax = sns.stripplot(data=df,x='Species',y='PetalLengthCm',jitter=True,edgecolor='gray')
```

```python
sns.violinplot(data=df,x="Species",y="PetalLengthCm",size=5)
```

```python
sns.FacetGrid(df,hue='Species',size=5)\
   .map(sns.kdeplot,'PetalLengthCm')\
   .add_legend()
```

```python
sns.pairplot(df.drop('Id',axis=1),hue='Species')
```

```python
sns.pairplot(df.drop('Id',axis=1),hue='Species',diag_kind='kde')
```

```python
df.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```
